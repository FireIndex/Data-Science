### Q6:- Write a program that will tell the number of dogs and chicken are there when the user will provide the value of total heads and legs.

For example:
Input:
heads -> 4
legs -> 12
<br>
Output:
dogs -> 2
chicken -> 2





````python
# Write your code here
heads = 4 # 5
legs = 12 # 18

''' 
# using put and test method
for let_dog_head in range(heads):
    chicken_head = heads - let_dog_head
    total_let_leg = 4*let_dog_head + 2*chicken_head

    if total_let_leg == legs:
        print('dogs ->', let_dog_head, 'chicken ->', chicken_head)
        # break
'''

# chicken + dogs = heads
# chicken = heads - dogs ... eq (1)

# 2*chicken + 4*dogs = legs
# ... put eq (1) in above eq
# 2(heads - dogs) + 4*dogs = legs
# 2*head - 2*dogs + 4*dogs = legs
# 2*dogs = legs - 2*heads ... eq (2)

dogs = (legs - 2*heads)/2
chicken = heads - dogs
if dogs >= 0 and chicken >= 0:
    print('dogs ->', dogs, 'chicken ->', chicken)
````
